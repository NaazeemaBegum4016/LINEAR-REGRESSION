```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error ,r2_score
```

```python
data ={
    "temperature" : [20,25,32,35,40,45,50,55,60],
    "sales" : [200,300,350,400,500,600,700,800,900]
    }
```

```python
df=pd.DataFrame(data)
```

```python
df
```

```python
x=df["temperature"]
y=df["sales"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model=LinearRegression()
```

```python
model.fit(x_train.to_frame(),y_train)
```

```python
y_pred=model.predict(x_test.to_frame())
```

```python
mse=mean_squared_error(y_test,y_pred)
r2_score=r2_score(y_test,y_pred)
```

```python
print(mse)
print(r2_score)
```

```python
new_temp=pd.DataFrame({"temperature":[65]})
new_sales=model.predict(new_temp)
print(new_sales)
```
